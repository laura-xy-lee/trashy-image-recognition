```python
import requests

from bs4 import BeautifulSoup

# Get html from webpage
r = requests.get("http://image-net.org/challenges/LSVRC/2014/browse-synsets")

# Extract html from main div
soup = BeautifulSoup(r.text)
result = soup.find("div", {"id" : "main"})

# Get all 1000 classifications
classification_dict = {}
classes = []
get = False

for classification in result.findAll('a'):
    text = classification.text
    
    if text == 'kit fox, Vulpes macrotis':
        get = True
            
    if get:
        classes.append(classification.text)
```

```python
# Remove animals

# Phase 1
classes_2 = ['dial telephone, dial phone',
             'maze, labyrinth']
get = False

for text in classes:
    if text == 'revolver, six-gun, six-shooter':
        get = True
            
    if get:
        classes_2.append(text)

# Phase 2
classes_2.remove("ant, emmet, pismire") 
classes_2.remove("starfish, sea star")
classes_2.remove("chambered nautilus, pearly nautilus, nautilus")

# Phase 3
classes_3 = []
get = True

for text in classes_2:
    if text == 'cock':
        get = False
    elif text == 'whistle':
        get = True
            
    if get:
        classes_3.append(text)

# Phase 4
classes_4 = []
get = True

for text in classes_3:
    if text == "spider web, spider's web":
        get = False
    elif text == "iron, smoothing iron":
        get = True
            
    if get:
        classes_4.append(text)

classes_4.remove("brown bear, bruin, Ursus arctos") 
classes_4.remove("skunk, polecat, wood pussy") 
classes_4.remove("meerkat, mierkat") 

# Phase 5
classes_5 = []
get = True

for text in classes_4:
    if text == "dumbbell":
        get = False
        
    if get:
        classes_5.append(text)
                
classes_5.remove('leatherback turtle, leatherback, leathery turtle, Dermochelys coriacea')
classes_5.remove('snow leopard, ounce, Panthera uncia')

# Classify animals

animals = set(classes) - set(classes_5)
classification_dict["animals"] = animals
```
